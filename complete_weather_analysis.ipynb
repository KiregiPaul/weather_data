below is the temparature trends for the various areas that is Nairobi Nyeri and Mombasa.

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df['year-month'] = df['datetime'].dt.to_period('M')

monthly_data0 = df.groupby('year-month').mean('temp')
monthly_data0['temp'].plot(kind='line')
plt.title('average temp by months in Nairobi')
plt.xlabel('Month')
plt.ylabel('temp')
plt.show()
```

# insights
The city of Nairobi experiences higher temparatures during the months of January, February and March and experiences quite low temperatures  during the months of June July and August. The minimum temperatures in Nairobo have gone to as low as 15.8 and a maximum temperature going to as high as 24.2
